# KMeans实现图像分割

```python
#图像加载与展示
import numpy as np
from skimage import io
img = io.imread('1.jpg')
```

```python
from matplotlib import pyplot as plt
plt.imshow(img)
```

```python
#查看图片存储后的数据类型与维度
print(type(img))
print(img.shape)
```

```python
#查看数据内容
print(img)
```

```python
#维度存储
img_width = img.shape[1]
img_height = img.shape[0]
print(img_width,img_height)
```

```python
#数据预处理 维度转化
img_data = img.reshape(-1,3)
print(img.shape,img_data.shape)
```

```python
#X赋值
X = img_data
```

```python
#模型建立
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3,random_state=0)
print(model)
```

```python
#模型训练
model.fit(X)
```

```python
#结果预测
label = model.predict(X)
print(label)
```

```python
import pandas as pd
print(pd.Series(label).value_counts())
```

```python
#确认预测结果数据量
print(label.shape)
```

```python
#结果数据维度恢复
label = label.reshape([img_height,img_width])
print(label.shape)
```

```python
#数据灰度化处理
# label = 1/(1+label)
print(label)
```

```python
print(label.shape)
```

```python
#结果可视化
plt.imshow(label)
```

```python
#图片存储
# label=label.astype(np.uint8)
io.imsave('result_k3.png',label)
```

```python
model_5 = KMeans(n_clusters=5,random_state=0)
#模型训练
model_5.fit(X)
#结果预测
label_5 = model_5.predict(X)
#结果数据维度恢复
label_5 = label_5.reshape([img_height,img_width])
#数据灰度化处理
# label_5 = 1/(1+label_5)
#结果可视化
plt.imshow(label_5)
#图片存储
# label_5=label_5.astype(np.uint8)
io.imsave('result_k5.png',label_5)
```
