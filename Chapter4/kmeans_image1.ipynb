```python
#图像加载
import numpy as np
# import skimage.io
from skimage import io
img=io.imread('output_9.png')
```

```python
from matplotlib import pyplot as plt
plt.imshow(img)
```

```python
#查看图片存储后的数据类型与维度
print(type(img))
print(img.shape)
```

```python
#打印数据内容
print(img)
```

```python
#维度存储
img_width=img.shape[1]
img_height=img.shape[0]
print(img_width,img_height)
```

```python
#数据预处理 维度转化
img_data=img.reshape(-1,4)
print(img_data,img_data.shape)
```

```python
#X赋值
X=img_data
print(X)
```

```python
#模型建立
from sklearn.cluster import KMeans
model=KMeans()
print(model)
```

```python
#模型训练
model.fit(X)
```

```python
#结果预测
label=model.predict(X)
print(label)
```

```python
import pandas as pd
print(pd.Series(label).value_counts())
```
